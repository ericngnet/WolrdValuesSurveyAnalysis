```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

import pickle
import pyreadstat

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_raw, meta = pyreadstat.read_dta('WV6_Data_Stata_v20180912.dta')
df_raw.head()
```

```python
df_raw.shape
```

```python
df_raw = df_raw.rename_axis('ID').reset_index()
```

```python
cow = pd.read_excel('C:/Users/Eric/Documents/Udacity/WordValueSurvey/COW.xlsx')
```

```python
cow.head()
```

```python
cow.rename(columns ={'Code': 'COW'}, inplace = True)
```

```python
df_raw = pd.merge(df_raw, cow,how="inner", on='COW')
```

```python
df_raw.rename(columns = {'Name': "Country"}, inplace = True)
df_raw["Country"].replace('United States of America', 'USA', inplace = True)
df_raw["Country"] = df_raw["Country"].str.strip()
print(df_raw.shape)
df_raw.head()
```

```python
df = df_raw.copy()
```

```python
df.shape
```

```python
df = df[df['RESEMAVAL'] > 0.001]
df = df[df['SACSECVAL'] > 0.001]
df.shape
```

```python
pd.set_option('display.max_rows', None)
df.isnull().sum()
```

```python
df_gp =  df[['ID','DEFIANCE','DISBELIEF', 'RELATIVISM','SCEPTICISM','AUTONOMY','EQUALITY','CHOICE','VOICE','SACSECVAL','RESEMAVAL']]
df_gp = df_gp.set_index('ID')
df_gp.head()
```

```python
plt.figure(figsize = (12, 9))
s = sns.heatmap(df_gp.corr(),
               annot = True, 
               cmap = 'RdBu',
               vmin = -1, 
               vmax = 1)
s.set_yticklabels(s.get_yticklabels(), rotation = 0, fontsize = 12)
s.set_xticklabels(s.get_xticklabels(), rotation = 90, fontsize = 12)
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize = (12, 9))
plt.scatter(df_gp.iloc[:,8], df_gp.iloc[:,9])
plt.xlabel('SACSECVAL')
plt.ylabel('RESEMAVAL')
plt.title('Visualization of raw data')
```

```python
sorted(df_gp['SACSECVAL'].unique().tolist())
```

```python
df_gp.head()
```

```python
df_8 = df_gp.iloc[:,:8]
```

```python
df_8.head()
```

```python
scaler = StandardScaler()
df_8_std = scaler.fit_transform(df_8)
```

```python
wcss = []
for i in range(1,12):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(df_8_std)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.figure(figsize = (10,8))
plt.plot(range(1, 12), wcss, marker = 'o', linestyle = '--')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.title('K-means Clustering')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 5, init = 'k-means++', random_state = 42)
```

```python
kmeans.fit(df_8_std)
```

```python
df_8_kmeans = df_8.copy()
df_8_kmeans['Values Groups'] = kmeans.labels_
df_8_kmeans.head()
```

```python
df_gp['Labels'] = df_8_kmeans['Values Groups']
```

```python
x_axis = df_gp['SACSECVAL']
y_axis = df_gp['RESEMAVAL']
plt.figure(figsize = (12, 10))
sns.scatterplot(x_axis, y_axis, hue = df_gp['Labels'], palette = ['red','orange','black','green','dodgerblue'])
plt.title('Values Groups')
plt.show()
```

```python
pca = PCA()
pca.fit(df_8_std)
PCA (copy = True, iterated_power = 'auto', n_components = None, random_state = None, svd_solver = 'auto',\
    tol=0.0, whiten = False)
```

```python
pca.explained_variance_ratio_
```

```python
# Plot the cumulative variance explained by total number of components.
# On this graph we choose the subset of components we want to keep. 
# Generally, we want to keep around 80 % of the explained variance.
plt.figure(figsize = (12,9))
plt.plot(range(1,9), pca.explained_variance_ratio_.cumsum(), marker = 'o', linestyle = '--')
plt.title('Explained Variance by Components')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
```

```python
pca = PCA(n_components = 6)
pca.fit(df_8_std)
```

```python
pca.components_
```

```python
df_pca_comp = pd.DataFrame(data = pca.components_,
                          columns = df_8.columns.values,
                          index = ['Component 1','Component 2','Component 3','Component 4','Component 5','Component 6'])
df_pca_comp
```

```python
sns.heatmap(df_pca_comp,
           vmin = -1,
           vmax = 1,
           cmap = 'RdBu',
           annot = True)
plt.yticks([0,1,2,3,4,5],
           ['Component 1','Component 2','Component 3','Component 4','Component 5','Component 6'],
           rotation = 45, fontsize = 9);
```

```python
scores_pca = pca.transform(df_8_std)
```

```python
wcss = []
for i in range(1,11):
    kmeans_pca = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans_pca.fit(scores_pca)
    wcss.append(kmeans_pca.inertia_)
```

```python
# Plot the Within Cluster Sum of Squares for the K-means PCA model. Here we make a decission about the number of clusters.
# Again it looks like four is the best option.
plt.figure(figsize = (10,8))
plt.plot(range(1, 11), wcss, marker = 'o', linestyle = '--')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.title('K-means with PCA Clustering')
plt.show()
```

```python
kmeans_pca = KMeans(n_clusters = 4, init = 'k-means++', random_state = 42)
kmeans_pca.fit(scores_pca)
```

```python
print(df_gp.shape)
print(df.shape)
```

```python
# We create a new data frame with the original features and add the PCA scores and assigned clusters.
df_gp['ID'], df_gp['Country'] = df['ID'], df['Country']
df_gp_pca_kmeans = pd.concat([df_gp.reset_index(drop = True), pd.DataFrame(scores_pca)], axis = 1)
df_gp_pca_kmeans.columns.values[-6: ] = ['Component 1', 'Component 2', 'Component 3','Component 4','Component 5','Component 6']
# The last column we add contains the pca k-means clustering labels.
df_gp_pca_kmeans['Values Groups'] = kmeans_pca.labels_
df_gp_pca_kmeans.head()
```

```python
df_gp_pca_kmeans['Values Groups'] = df_gp_pca_kmeans['Values Groups'].map({0:'1', 
                                                          1:'2',
                                                          2:'3', 
                                                          3:'4'})
df_gp_pca_kmeans.head()
```

```python
df_gp_pca_kmeans.drop(columns = 'Labels', inplace = True)
```
